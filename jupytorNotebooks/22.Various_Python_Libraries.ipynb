## time

- ``time.time()``  
    - ``UTC``(Universal Time Coordinated 협정 세계 표준시)를 사용하여 현재 시간을 실수 형태로 돌려주는 함수  
    - 1970년 1월 1일 0시 0분 0초를 기준으로 지난 시간을 초 단위로 돌려준다
    
    
- ``time.localtime``  
    - ``time.time()``이 돌려준 실수 값을 사용해서 연도, 월, 일, 시, 분, 초, ... 의 형태로 바꾸어 주는 함수
    
    
    
- ``time.strftime``('출력할 형식 포맷 코드', ``time.localtime(time.time())``) - string from time

```
%x  날짜를 출력, 현재 시간대 기준
%c  날짜, 요일, 시간을 출력, 현재 시간대 기준
%Y  네 자릿수 연도
%m  숫자 월
%d  일
%H  시간(24시간)
%M  분
%S  초
%z  시간대
%a  요일 줄임말
%A  요일
%b  월 줄임말
%B  월
%I  시간(12시간)
%p  AM, PM
```

````python
import time

time.time()
````

````python
time.localtime(time.time())
````

````python
time.strftime('%x', time.localtime(time.time()))
````

````python
time.strftime('%c', time.localtime(time.time()))
````

````python
time.strftime('%Y-%m-%d-%H:%M:%S %A', time.localtime(time.time()))
````

### image 를 읽어서 display 

- ``Image.open()`` : image file open
    

- internet 상의 image 는 byte code 로 전송되므로 ``BytesIO`` 를 이용하여 binary 로 변환  

````python
import requests
from PIL import Image
from io import BytesIO

url = "https://c.files.bbci.co.uk/12A9B/production/_111434467_gettyimages-1143489763.jpg"
````

````python
response = requests.get(url)
Image.open(BytesIO(response.content)).resize((300, 200))
````

````python
img = Image.open("images/dog.0.jpg")
img = img.resize((300, 200))
img
````

````python
print("Format: {0}\nSize: {1}\nMode: {2}".format(img.format, img.size, img.mode))
````

````python
path = "images/dog.0.jpg"
url = "https://c.files.bbci.co.uk/12A9B/production/_111434467_gettyimages-1143489763.jpg"

def image_load(file_path):
    if 'http' in file_path:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content)).resize((300, 200))
    else:
        img = Image.open("images/dog.0.jpg").resize((300, 200))
        
    return img
        
img = image_load(url)
img
````

# Python 을 이용하여 zipfile handling 하는 방법

### zipfile
- zip archive object 를 생성하고 개별 file 을 add

### shutl
- 전체 directory 를 빠르게 compress

````python
# sample file 2 개 생성
f = open('new_file_1.txt', 'w')
f.write('이것은 첫번째 테스트 file 입니다.')
f.close()

f = open('new_file_2.txt', 'w')
f.write('두번째 테스트 file 입니다')
f.close()
````

````python
import zipfile

# empty zipfile object 생성
comp_file = zipfile.ZipFile('new_compressed_file.zip', 'w')
````

````python
# empty object 에 압축할 file 추가
comp_file.write('new_file_1.txt')
comp_file.write('new_file_2.txt')
````

````python
comp_file.close()
````

## Extract from zipfile

````python
# 읽어들일 zipfile object 생성
zip_object = zipfile.ZipFile('new_compressed_file.zip', 'r')

# 새로운 directory 에 모든 file 압축 해제
zip_object.extractall('unzipped_content')
````

````python
# 개별 file 해제 (file name, 저장 directory)
zip_object.extract('new_file_1.txt', 'unzipped_content_2')
````

## shutil Library

````python
import shutil

# make_archive (압축 file name, format, source directory)
shutil.make_archive('zipped_content3', 'zip', 'unzipped_content')
````

````python
# archived file 압축 해제 : unpack(압축해제할 filename, target directory, format)
shutil.unpack_archive('zipped_content3.zip', 'unzipped_content_3', 'zip')
````
